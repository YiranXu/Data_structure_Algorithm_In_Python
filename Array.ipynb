## My notes when studying array in Python

Array is a data structure that is a collection of the same type.


In Python, a list or tuple is an array of object references. It stores a consecutive sequence of memory addresses at which the elements of the sequence reside. 


Becuase lists in python stores the references, lists store multiple types of elements.

#### Compact Array
When the array is storing the bits that represent the primary data (eg. characters in string), we call it a compact array.For instance, Strings are represented using an array of characters (not an array of references). 

Compact arrays have multiple advantages:
* Uses much lower space/ memory because it requires no overhead of extra storage of the sequence of memory references
* primary data are stored consecutively in memory, which is not the case for a referential structure. 
    Because of the workings of the cache and memory hierachies of computers, it is often advantageous to have data stored in memory near other data that might be used in the same computations.

For string, each character is represented using a Unicode character and each unicode character is 16 bits(i.e. 2 bytes). For referential structures, the number of bits used to store the memory address of each element is 64-bits per address.

The module "array" in python is compact array. It can store basic values: characters, integers, floating point numbers. https://docs.python.org/3/library/array.html

Users need to defind *type code* to let the interpreter to determine how many bits are needed per element of the array.

Compact arrays of user-defined data types can be created with another lower-level support module "ctypes"

### Coding exercise
Write Python version of leetcode learn section "Array" whose code is in JAVA.

```python
dvd_collection=[None]*15 
class DVD():
    def  __init__(self,name,release_year,director):
        self.name=name
        self.release_year=release_year
        self.director=director
    def to_string(self):
        print(f"{self.name}, directed by {self.director}, released in {self.release_year}")
```

##### Writing items to an array

```python
avengers_DVD=DVD('The Avengers',2012,'Joss Whedon')
avengers_DVD.to_string()
```

Since we have initialized a list, to insert into the eighth place, we actually want to replace the element to new element (make it point to a new object). The insert method would actually insert by adding a new element. 

```python
#insert into the eighth place （replace) 
dvd_collection[7]=avengers_DVD
```

```python
incredibles_DVD = DVD("The Incredibles", 2004, "Brad Bird")
findingDory_DVD = DVD("Finding Dory", 2016, "Andrew Stanton")
lionKing_DVD = DVD("The Lion King", 2019, "Jon Favreau")
```

```python
#Put "The Incredibles" into the 4th place: index 3.
dvd_collection[3] = incredibles_DVD

#Put "Finding Dory" into the 10th place: index 9.
dvd_collection[9] = findingDory_DVD

#Put "The Lion King" into the 3rd place: index 2.
dvd_collection[2] = lionKing_DVD
```

```python
#Overwritten
starWars_DVD = DVD("Star Wars", 1977, "George Lucas");
dvd_collection[3] = starWars_DVD
```

```python
dvd_collection[3].to_string()
```

```python
#Reading Items from an Array
print(dvd_collection[7])
print(dvd_collection[10])
```

#### Writing Items into an Array with a Loop
we're going to create an Array of ints and put the first 10 square numbers into it.

```python
square_numbers=[i** 2 for i in range(10)]
```

#### Reading Items from an Array with a Loop
Using "for each" loop without explicitly specify actual index numbers is more elegant.

```python
for square in square_numbers:
    print(square)
```

## Inserting Items into an array
1. Insert at the end of an Array
using append method in Python --> In the worst case, it requires $\Omega (n)$ time because the underlying array is resized, but it uses $O(n)$ time in the amortized sense.
2. Insert at the start of an Array 
The most expensive insert, requiring $O(N)$ liner time per operation where N is the length of the array
3. Insert anywhere in the Array
we need to shift all the elements from the index onwards one position to the right
inserting at index $k$ ---> shift $n-k-1$ elements ---> $O(n-k-1)$
